```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.arange(10)
y_data = (x_data-5)**2
print(x_data,y_data)
```

```python
plt.style.use('seaborn')
plt.plot(x_data,y_data)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

Gradient algo

```python
X = 0
lr = 0.1
error = []
# 50 steps int the downhill direction
plt.plot(x_data,y_data)

for i in range(50):
    grad = 2*(X-5)
    X = X - lr*grad
    Y = (X-5)**2
    error.append(Y)
    plt.scatter(X,Y)
    print(X)
    
```

```python
plt.plot(error)
plt.show()
```
